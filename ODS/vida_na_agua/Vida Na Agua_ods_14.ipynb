# Doenças Tropicais Negligenciadas ou DTN's

As Doenças Tropicais Negligenciadas (DTNs) ameaçam mais de 1,7 bilhão de pessoas que vivem nas comunidades mais pobres e marginalizadas do mundo. Causadas por uma variedade de patógenos, incluindo vírus, bactérias, parasitas, fungos e toxinas, esses agravos cegam, incapacitam e desfiguram as pessoas, tirando não apenas sua saúde, mas também suas chances de permanecer na escola, de ganhar a vida ou mesmo de ser aceito por sua família ou comunidade.

Essas enfermidades também apresentam indicadores inaceitáveis e investimentos reduzidos nas áreas de pesquisa, produção de medicamentos e em seu controle.

Hanseníase, dengue, leishmaniose, esquistossomose, raiva humana transmitida por cães, escabiose (sarna), doença de Chagas, parasitoses intestinais e tracoma são algumas das mais de 20 patologias presentes na região – onde também são conhecidas como doenças infecciosas negligenciadas. Juntas, causam entre 500 mil e 1 milhão de óbitos anualmente.

```python
import geopandas as gpd

# Carregar o arquivo shapefile para um GeoDataFrame
gdf = gpd.read_file('2023/IDSC-BR_2023_Cidades-Estudadas_Poligonos.shp')

# Selecionar as colunas desejadas
ods14 = gdf[['CD_MUN', 'AREA_KM2', 'NOME_MUNIC', 'UF', 'IDSC-BR', 'Classifica','ODS_14_pt','geometry']]

# Filtrar os dados
vida_na_agua = ods14[ods14['ODS_14_pt'] != 0]

# Exibir uma amostra dos dados
vida_na_agua.sample(5)
```

Deixando em ordem por pontuação da meta selecionada

```python
vida_na_agua = vida_na_agua.sort_values(by='ODS_14_pt', ascending=False)
#vida_na_agua['posicao_ods14'] = range(1, len(vida_na_agua) + 1)
vida_na_agua
```

```python
santos = vida_na_agua[vida_na_agua['NOME_MUNIC'] == 'Santos']
santos = santos['ODS_14_pt'].iloc[0]
print(f'A pontuação da cidade de Santos é de {santos} pontos na categoria Vida Na Água segundo a ICSC-BR em 2023')
```

# Adicionando classificação de acordo com a ODS 14

Já temos a classificação geral, agora quero separar as cidades em categorias de acordo com os limiares da ICSC-BR

```python
limiar_vermelho = 40
limiar_verde = 70

def limiares(valor):
    if valor <= limiar_vermelho:
        return 'Vermelho'
    elif valor <= limiar_verde:
        return 'Laranja'
    elif valor < 100:  
        return 'Verde'
    else:
        return 'Alvo'

vida_na_agua['categoria_ods14'] = vida_na_agua['ODS_14_pt'].apply(lambda x: limiares(x))
vida_na_agua.sample(5)
```

# Porque apenas longo curso ?

 "Longo Curso é aquele realizado entre portos de diferentes países, em oposição ao transporte marítimo de cabota- gem, realizado entre os portos localizados dentro do território nacional de um país."

```python
longo_curso = atracacao[atracacao['Tipo de Navegação da Atracação'] == 'Longo Curso']
longo_curso.sample(3)
```

# Atribuindo Soma De Operações à Portos e Portos à Cidades

```python
# 
soma_por_operacao = longo_curso.groupby(['SGUF', 'Município'])['Tipo de Operação'].count().reset_index()

# Renomear a coluna de contagem para 'Quantidade de Operações'
soma_por_operacao.rename(columns={'Município':'NOME_MUNIC','SGUF':'UF','Tipo de Operação': 'Quantidade de Operações'}, inplace=True)

# Exibir a soma por id de operação
soma_por_operacao
```

# Agora vou mergear com o resto do dataframe

```python
vida_na_agua = pd.merge(vida_na_agua, soma_por_operacao[['NOME_MUNIC', 'UF','Quantidade de Operações']], on=['NOME_MUNIC', 'UF'], how='left')
vida_na_agua
```

```python
portos_vida_na_agua = vida_na_agua[vida_na_agua['Quantidade de Operações'].notna()]
portos_vida_na_agua.sample(3)
```
